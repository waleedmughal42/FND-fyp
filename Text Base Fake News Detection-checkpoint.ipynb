```python
#function to plot the confusion matrix
#this function prints and plot the confusion matrix

def plot_confusion_matrix(cm,classes,normalize=False,title='Confusion matrix',cmap=plt.cm.Blues):
    plt.imshow(cm,interpolation='nearest',camp=camp)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm=cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normailze confusion matrix")
    else:
        print("confusion matrix without Normalization")
        
    thresh = cm.max / 2.
    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j], 
                 horizentalalignment = "center", 
                 color = "white" if cm[i,j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicated label')
```

```python
#build classifer using navie bayes
#Navie Bayes classifer for Multinomial model

nb_pipeline = Pipeline([('NBTV', tfidf_vectorizer), ('nb_clf', MultinomialNB())])
```

```python
#fit navie byes classifer according to X and Y

nb_pipeline.fit(X_train, y_train)
```

```python
#perform classifiation  on an array of test vector X

predicted_nbt = nb_pipeline.predict(X_test)
```

```python
score = metrics.accuracy_score(y_test, predicted_nbt)
print(f'Accuracy: {round(score*100,2)}%')
```

```python
cm = metrics.confusion_matrix(y_test, predicted_nbt, labels=['FAKE', 'REAL'])
plot_confusion_matrix(cm, classes = ['FAKE', 'REAL'])
```
